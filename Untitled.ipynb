```python
# library
import matplotlib.pyplot as plt
 
```

```python
# Data
names='Handgun', 'Blunt Object', 'Knife', 'Firearm', 'Unknown', 'Shotgun', 'Rifle', 'Strangulation', 'Fire', 'Suffocation', 'Gun', 'Drugs', 'Drowning', 'Poison', 'Explosives', 'Fall'
size=[116205, 23945, 29118, 28077, 16190, 7186, 6874, 1905, 1776, 1638, 1495, 981, 319, 184, 82, 47]
```

```python
# create a figure and set different background
fig = plt.figure()
fig.patch.set_facecolor('black')
```

```python
plt.rcParams['text.color'] = 'black'
```

```python
my_circle=plt.Circle( (0,0), 0.7, color='white')
```

```python
plt.pie(size, labels=names, colors=['red', 'blue', 'purple','green'])
p=plt.gcf()
p.gca().add_artist(my_circle)
```

```python

plt.show()
plt.savefig('Plot3.png')
```
