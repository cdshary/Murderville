```python
import matplotlib.pyplot as plt
from math import atan2,degrees
import numpy as np
```

```python
year=[2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]
```

```python
Louisiana=[558, 511, 577, 601, 586, 433, 516, 607, 396, 520, 493, 509, 511, 478, 452]
```

```python
USA_avg=[287.67, 309.86, 318.98, 323.76, 318.29, 330.12, 338.73, 339.27, 305.78, 310.59, 296.49, 289.33, 294.76, 283.24, 281.00]
```

```python
plt.plot(year, Louisiana, color='g', label='Louisiana')
plt.plot(year, USA_avg, color='orange', label='National Average')
plt.xlabel('Years')
plt.ylabel('Murders')
plt.title('Louisiana vs USA Average Murders per 100,000')
plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
fig = plt.figure()
plt.savefig('plot2.png')
```
